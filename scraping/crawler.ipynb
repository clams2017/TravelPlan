```python
# with open('gurutabi_fukuroda_waterfall.htm') as fDetail:
with open('gurutabi_ueno_zoo.htm') as fDetail:
    s = BeautifulSoup(fDetail, 'html.parser')
# print (s)
```

```python
name = s.find('h1', class_='main-content__headline').string.strip()
print (name)
```

```python
mt40s=s.findAll('div', class_='mt40')
for mt40 in mt40s:
    sections = mt40.findAll('h2', class_='item-section-heading')
    for section in sections:
        if section.string == '概要':
            description = mt40.find('p').string
            break;
print(description)
```

```python
access = s.find('div', class_='access-map mt40')
lat = access.find('input', id='latitudeValue')
lon = access.find('input', id='longitudeValue')
print(lat.get('value'), lon.get('value'))
```

```python
breadcrumbs=s.find('ol', class_='breadcrumb__list').findAll('li')
for breadcrumb in breadcrumbs:
    a = breadcrumb.find('a')
    if not a:
        continue
    url = a.get('href')
    paths = url.split('/')
    if paths[-2].startswith('gl'):
        gl=a.string
    if paths[-2].startswith('gm'):
        gm=a.string
    if paths[-2].startswith('gs'):
        gs=a.string
print (gl)
print (gm)
print (gs)
```

```python
bxslider=s.find('ul', class_='bxslider')
images=bxslider.findAll('img')
imageArray=[]
for image in images:
    imageArray.append(image.get('src'))
print(imageArray)
```

```python
bxslider=s.find('ul', class_='bxslider')
image=bxslider.find('img').get('src')
print(image)
```

```python
access = s.find('div', class_='access-map mt40')
accessText = access.find('p', class_='access-map__txt')
accessArray = []
for str in accessText.strings:
    accessArray.append(str.strip())
accessString = '\n'.join(accessArray).strip()
print(accessString)
```
